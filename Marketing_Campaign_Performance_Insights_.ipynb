```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#Campaign Performance:

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Acquisition_Cost',y='ROI',hue='Campaign_Type',data=df,palette='viridis')
plt.title('Relationship between Acquisition_Cost and ROI')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.show()
```

```python
avg_conversation_rate = df.groupby('Campaign_Type')['Conversion_Rate'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(hue='Campaign_Type',x='Campaign_Type',y='Conversion_Rate',data =avg_conversation_rate ,palette="viridis",legend=True)

plt.xlabel('Campaign_Type')
plt.ylabel('Average Conversation Rate')
plt.title('average Conversion_Rate for different Channel_Used')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Engagement_Score',hue='Campaign_Type',y='Campaign_Type',data=df,palette='viridis',legend=False)
plt.title('Engagement_Score Distribution by Campaign_Type')
plt.xlabel('Engagement_Score')
plt.ylabel('Campaign_Type')
plt.show()
```

```python
avg_profit_rate = df.groupby('Company')['ROI'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(hue='Company',x='Company',y='ROI',data =avg_profit_rate ,palette="viridis")

plt.xlabel('Company')
plt.ylabel('Return on Investment')
plt.title(' profitable of campaigns conducted by different companies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['Engagement_Score','Conversion_Rate']

corr = df[numeric_cols].corr()
plt.figure(figsize=(10,7))

sns.heatmap(
    corr,
    annot=True,
    cmap='viridis',
    fmt='.2f',
    linewidths=1,
)

plt.title('correlation between Engagement_Score and Conversion_Rate ')
plt.show()
```

#Customer Segmentation:

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Target_Audience',hue='Target_Audience',data=df,palette='viridis')
plt.title('Target Audience Distribution')
plt.xlabel('Target_Audience')
plt.ylabel('Count of Target Audience Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Language and Customer_Segment, take mean of Conversion_Rate
grouped = df.groupby(['Language','Customer_Segment'])['Conversion_Rate'].mean().reset_index()

# For each Language, get the Customer_Segment with the highest Conversion_Rate
highest_per_language = grouped.loc[grouped.groupby('Language')['Conversion_Rate'].idxmax()]
print(highest_per_language)


import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(
    highest_per_language['Language'],
    highest_per_language['Conversion_Rate'],
    color=plt.cm.viridis(range(len(highest_per_language)))
)

# Annotate Customer_Segment on the bars
for index, value in enumerate(highest_per_language['Conversion_Rate']):
    plt.text(value + 0.01, index, highest_per_language['Customer_Segment'].iloc[index])

plt.title('Highest Conversion Rate Customer Segment for Each Language')
plt.xlabel('Conversion Rate')
plt.ylabel('Language')
plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.boxplot(
    x='Acquisition_Cost',
    hue='Channel_Used',
    y='Channel_Used',
    data=df,
    palette='viridis'
)

plt.title('Acquisition_cost Distribution by Channel_Used')
plt.xlabel('Acquisition_Cost')
plt.ylabel('Channel_Used')
plt.show()
```

```python
highest_Conversion_Rate = df.groupby('Language')['Conversion_Rate'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.barh(
    highest_Conversion_Rate['Language'],
    highest_Conversion_Rate['Conversion_Rate'],
    color=plt.cm.viridis(range(len(highest_Conversion_Rate)))
)

plt.title('Average Conversion_Rate by Language')
plt.xlabel('Conversion Rate')
plt.ylabel('Language')
plt.show()
```

#Channel Effectiveness:

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Engagement_Score',y='Channel_Used',hue='Campaign_Type',data=df,palette='viridis')
plt.title('Engagement_Score Distribution by Channel_Used')
plt.xlabel('Engagement_Score')
plt.ylabel('Channel_Used')
plt.show()
```

```python
tot_ROI = df.groupby('Channel_Used')['ROI'].sum()

plt.figure(figsize=(8,8))
plt.pie(
    tot_ROI.values,
    labels=tot_ROI.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('total ROI across different Channels_Used')
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.scatterplot(x='Clicks',y='Impressions',hue='Channel_Used',data=df,palette='viridis')
plt.title(' relationship between Clicks and Impressions for each Channel_Used')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.show()
```

#Time-Based Analysis:

```python
plt.figure(figsize=(10,6))

plt.hist(df['Duration'], bins=10, edgecolor='black', color='orange')
plt.title('Distribution of Campaign Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
```

```python
g = sns.FacetGrid(df, col="Company", col_wrap=3, height=5, sharey=True)
g.map_dataframe(sns.lineplot, x="Date", y="Conversion_Rate", marker="o")
g.set_axis_labels("Date", "Conversion_Rate")
g.set_titles(col_template="{col_name}")
for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='Date', y='Engagement_Score',data=df , marker='o',linestyle = '--' ,color='skyblue')
plt.title('Engagement_Score Over Time')
plt.xlabel('Date')
plt.ylabel('Engagement_Score')
plt.show()
```

#Geographic Analysis:

```python
Highest_Acquisition_Cost = df.groupby('Location')['Acquisition_Cost'].max().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(data=Highest_Acquisition_Cost,x='Location',hue='Location',y='Acquisition_Cost',palette='viridis',legend=False)
plt.title('Highest Acquisition Cost for Each Location')
plt.xlabel('Location')
plt.ylabel('Highest Acquisition Cost')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Conversion_Rate',y='Target_Audience',hue='Location',data=df,palette='viridis')
plt.title('Conversion Rate Distribution by Target Audience')
plt.xlabel('Conversion Rate')
plt.ylabel('Target Audience')
plt.show()
```

```python
avg_ROI = df.groupby('Location')['ROI'].mean()

plt.figure(figsize=(10,7))
plt.pie(avg_ROI, labels=avg_ROI.index, autopct='%1.1f%%', startangle=90)

plt.title('Average Return on Investment by Location')
plt.show()
```
